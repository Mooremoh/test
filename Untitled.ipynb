```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Example data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Load an example image (replace 'your_image_path.png' with the actual image path)
img = plt.imread('map7.png')

# Create a scatter plot
plt.scatter(x, y, color='blue', marker='o', label='Scatter Points', s=50)

# Display the image using imshow
plt.imshow(img, extent=[min(x), max(x), min(y), max(y)], aspect='auto')

# Add labels and a title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot Overlaid on Image')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
df1=pd.read_csv('GrowLocations.csv')
df= df1.copy()
df1 = df1[(df1['Longitude'] >= -10.592) & (df1['Longitude'] <= 1.6848) &
        (df1['Latitude'] >= 50.681) & (df1['Latitude'] <= 57.985)]
x=df1[['Latitude']].values.tolist()
y=df1[['Longitude']].values.tolist()

# Create a scatter plot
plt.scatter(x, y, color='blue', marker='o', label='Scatter Points', s=150)
plt.imshow(img, aspect=[x,y])
# Add labels and a title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Scatter Plot of Coordinates')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
